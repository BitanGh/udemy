```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Travel.csv')
df.head()
```

Data Cleaning


```python
df.isnull().sum()
```

```python
df['MaritalStatus'].value_counts()
```

```python
df['ProductPitched'].value_counts()
```

```python
df['Occupation'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
#Replace Fe male with Female in Gender
df['Gender'] = df['Gender'].replace('Fe Male','Female')
df['Gender'].value_counts()
```

```python
df['MaritalStatus'] = df['MaritalStatus'].replace('Single','Unmarried')
df['MaritalStatus'].value_counts()
```

```python
#want to have the exact percentage of NaN value from each categry
df.isnull().sum()/len(df)*100
```

```python
#Filling Null values with median and mode

#Age
df['Age'] = df['Age'].fillna(df['Age'].median())

#Type of contract with mode
df['TypeofContact'] = df['TypeofContact'].fillna(df['TypeofContact'].mode()[0])

#Duration of Pitch
df['DurationOfPitch'] = df['DurationOfPitch'].fillna(df['DurationOfPitch'].median())

#Number of follow ups with mode
df['NumberOfFollowups'] = df['NumberOfFollowups'].fillna(df['NumberOfFollowups'].mode()[0])
#Preferred property star
df['PreferredPropertyStar'] = df['PreferredPropertyStar'].fillna(df['PreferredPropertyStar'].mode()[0])
#No of trips with median
df['NumberOfTrips'] = df['NumberOfTrips'].fillna(df['NumberOfTrips'].median())
#No of children visiting with mode
df['NumberOfChildrenVisiting'] = df['NumberOfChildrenVisiting'].fillna(df['NumberOfChildrenVisiting'].mode()[0])
#Monthly income with median
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(df['MonthlyIncome'].median())
```

```python
df.isnull().sum()
```

Featured Engineering


```python
#drop customer id
df.drop('CustomerID',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['TotalVisiting'] = df['NumberOfChildrenVisiting'] + df['NumberOfPersonVisiting']
df.drop(['NumberOfChildrenVisiting','NumberOfPersonVisiting'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#get all the numeric features
numeric_features = [i for i in df.columns if df[i].dtype != 'O']
len(numeric_features)
```

```python
#categorical Features
categorical_features = [i for i in df.columns if df[i].dtype == 'O']
len(categorical_features)
```

```python
#Getting Discrete categories
discrete_features = [i for i in numeric_features if len(df[i].unique()) <= 25]
len(discrete_features)
```

```python
#Continious Features
continious_features = [i for i in numeric_features if i not in discrete_features]
len(continious_features)
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X = df.drop(['ProdTaken'],axis=1)
y = df['ProdTaken']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_train.shape, X_test.shape
```

```python
categorical_features = X.select_dtypes(include='object').columns
numeric_features = X.select_dtypes(exclude='object').columns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
num_transformer = StandardScaler()
oh_transformer = OneHotEncoder(drop='first')
preprocessor = ColumnTransformer(
    [("OneHotEncoder",oh_transformer,categorical_features),
    ("StandardScaler",num_transformer,numeric_features)]
)
```

```python
#applying preprocessor for training data
X_train_processed = preprocessor.fit_transform(X_train)
pd.DataFrame(X_train_processed)
```

```python
X_test_processed = preprocessor.transform(X_test)
pd.DataFrame(X_test_processed)
```

```python
#RandomForest, metrices
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


```

```python
#Training
models = {
    "Random Forest": RandomForestClassifier()

}
for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train_processed,y_train)
```

```python
y_test_pred = model.predict(X_test_processed)
y_train_pred = model.predict(X_train_processed)
```

```python
model_train_accuracy = accuracy_score(y_train,y_train_pred)
model_test_accuracy = accuracy_score(y_test,y_test_pred)
```

```python
model_train_accuracy, model_test_accuracy
```

```python
#Hyperparameter
rf_params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2'],
    'bootstrap': [True, False]



}
```

```python
randomcv_models = [
     ('Random Forest',RandomForestClassifier(), rf_params)
]
```

```python
from sklearn.model_selection import RandomizedSearchCV
model_params = {}
for name, model, params in randomcv_models:
    random = RandomizedSearchCV(estimator=model,
                                param_distributions=params,
                                n_iter=100,
                                cv=3,
                                verbose=2,
                                n_jobs=-1)
    random.fit(X_train_processed,y_train)
    model_params[name] = random.best_params_
for model_name in model_params:
    print(f"Best parameters for {model_name}:")
    print(model_params[model_name])
    print("="*50)


```

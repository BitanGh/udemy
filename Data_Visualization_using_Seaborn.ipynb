```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="darkgrid")
dataset = sns.load_dataset("tips")
dataset
```

```python
sns.scatterplot(x="total_bill", y="tip", data=dataset)
plt.title("Scatter Plot")
plt.show()
```

```python
#Creating a LinePlot
sns.lineplot(x="total_bill", y="tip", data=dataset)
plt.title("Line Plot")
plt.show()
```

```python
#Creating a CategoricalPlot
sns.barplot(x="day", y="total_bill", data=dataset)
plt.title("Categorical Bar Plot")
plt.show()
```

```python
#Boxplot
sns.boxplot(x="day", y="total_bill", data=dataset)
plt.title("Box Plot")
plt.show()
```

```python
#Histogram plot
sns.histplot(dataset["total_bill"], bins= 10, kde=True)
plt.title("Histogram Plot")
```

```python
#PairPlot
sns.pairplot(dataset)
```

```python
#Heatmap
corr = dataset[["total_bill","tip","size"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
```
